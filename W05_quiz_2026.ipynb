## Question 3:

What combination of borough and year hard the largest number of homicide victims?

```python
con.sql("""
SELECT boroname, year, sum(NUM_VICTIM::INTEGER) AS homicide_count
FROM nyc_homicides
GROUP BY boroname, year
ORDER BY homicide_count DESC
LIMIT 1
""")
```

## Question 4:

Which subway station has the largest number of shooting victims within a 300 meter radius?


```python
con.sql("""
SELECT s.name, sum(h.NUM_VICTIM::INTEGER) AS homicide_count
FROM nyc_homicides h
JOIN nyc_subway_stations s
ON ST_DWithin(s.geom, h.geom, 300)
WHERE h.weapon = 'gun'
GROUP BY s.name
ORDER BY homicide_count DESC
LIMIT 2
""")
```

## Question 5

what is the most densely populated residential street in new york? (total population of intersecting census blocks divided by the total length of the street)

```python
con.sql("""
SELECT
    s.name,
    SUM(c.popn_total) / SUM(ST_Length(s.geom)) AS density
FROM nyc_streets s
JOIN nyc_census_blocks c ON ST_Intersects(s.geom, c.geom)
WHERE s.type = 'residential'
GROUP BY s.name
ORDER BY density DESC
LIMIT 1
""")
```

# Section 2

This section uses internet speed data from [Ookla](https://github.com/teamookla/ookla-open-data); take a minute to read the documentation of the dataset.

Create a table called ookla_nyc which reads the data from this url:

https://storage.googleapis.com/qm2/CASA0025/ookla_nyc.parquet




```python
con.sql("""
CREATE OR REPLACE TABLE ookla_nyc AS
SELECT * FROM read_parquet('https://storage.googleapis.com/qm2/CASA0025/ookla_nyc.parquet')
""")
con.sql("CREATE OR REPLACE TABLE ookla_nyc AS SELECT *, ST_GeomFromText(geometry) AS geom FROM ookla_nyc")
con.sql("SELECT * FROM ookla_nyc LIMIT 5").show()
con.sql("CREATE INDEX idx ON ookla_nyc USING RTREE(geom)")
```

## Question 6
How many tiles are contained within new york city neighbourhoods?

```python
con.sql("""
SELECT COUNT(DISTINCT(o.quadkey))
FROM ookla_nyc o
JOIN nyc_neighborhoods n
ON ST_Contains(n.geom, o.geom)
""")
```

## Question 7

Which neighbourhood has the highest average download speed? use 'within' rather than intersects, and ignore tiles with fewer than 3 tests.

```python
con.sql("""
SELECT n.name, avg(o.avg_d_kbps)
FROM ookla_nyc o
JOIN nyc_neighborhoods n
ON ST_Within(o.geom, n.geom)
WHERE o.tests >= 3
GROUP BY n.name
ORDER BY avg(o.avg_d_kbps) DESC
LIMIT 1
""")
```

## Question 8



What is the name of the subway station closest to the tile with the highest download speed?

```python
con.sql("""
SELECT s.name
FROM nyc_subway_stations s,
(SELECT geom FROM ookla_nyc ORDER BY avg_d_kbps DESC LIMIT 1) AS o
ORDER BY ST_Distance(s.geom, o.geom) ASC
LIMIT 1
""")
```

## Question 9

Which neighbourhood has the highest number of devices per capita? Use only spatial intersections.

```python
con.sql("""
SELECT n.name,
(SELECT SUM(devices) FROM ookla_nyc o WHERE ST_Intersects(n.geom, o.geom))/
(SELECT SUM(popn_total) FROM nyc_census_blocks c WHERE ST_Intersects(n.geom, c.geom))
AS devices_per_capita
FROM nyc_neighborhoods n
WHERE (SELECT SUM(popn_total) FROM nyc_census_blocks c WHERE ST_Intersects(n.geom, c.geom)) > 0
ORDER BY devices_per_capita DESC
LIMIT 1
""")
```

## Question 10


What is the average upload speed for tiles that are within 200 meters of L train stops and intersect with census blocks where the black population is above 50%?

```python
con.sql("""
SELECT avg(o.avg_u_kbps)
FROM ookla_nyc o
JOIN nyc_subway_stations s
    ON ST_DWithin(o.geom, s.geom, 200)
    AND s.routes LIKE '%L%'
JOIN nyc_census_blocks c
    ON ST_Intersects(o.geom, c.geom)
    AND c.POPN_TOTAL > 0
    AND (c.POPN_BLACK / c.POPN_TOTAL) > 0.5
""")
```
